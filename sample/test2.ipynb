```python
"""네이버 뉴스 기사 웹 크롤러 모듈"""
 
from bs4 import BeautifulSoup
import urllib.request
#입력 파일 명
open_file='data/news_url.txt'  
f = open(open_file, 'r')
lines = f.readlines()
f.close()

# 크롤링 함수, 단처리
def get_text(line):
 #   print(line)
    # 읽을 URL 입력 지점
    source_code_from_URL = urllib.request.urlopen(line) 
    print(line)
    soup = BeautifulSoup(source_code_from_URL, 'lxml', from_encoding='utf-8')
    text = ''
    # 파싱, append
    for item in soup.find_all('div', id='articleBodyContents'):
        text = text + str(item.find_all(text=True))
        print(text)
    return text

```

```python
# 메인 함수, 다처리
def crawler(lines):
    for i,line in enumerate(lines):      
        # 출력 파일 명
        out_file = 'data/navernews'+str(i)
        open_output_file = open(out_file, 'w', encoding='utf-8')
  #      print(line)
        result_text = get_text(line)
        
   #     print(result_text)
        # 파일 write
        open_output_file.write(result_text)
        open_output_file.close()
    
crawler(lines) 
```
